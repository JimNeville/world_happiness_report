# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

whr_all_years = pd.read_csv('../data/years_combined.csv')
whr_all_years.head(5)
```

```python
# Average happiness score by region by year

pd.pivot_table(data = whr_all_years, index='region', values='happiness_score').sort_values(by='happiness_score', ascending=False)
```

```python
# Average happiness score by region by year (sorted by 2017 desc)

pd.pivot_table(data=whr_all_years, index='region', columns='year', values='happiness_score').sort_values(by=2017, ascending=False)
```

```python
# Happies countries across 2015-2017

happiness_rank = whr_all_years.pivot_table(index='country', values='happiness_rank', aggfunc=np.median).sort_values(by='happiness_rank')

happiness_rank['rank'] = happiness_rank['happiness_rank'].rank(method='dense')
happiness_rank.drop('happiness_rank', inplace=True, axis=1)
happiness_rank.rename({'rank':'happiness_rank'}, axis=1)
happiness_rank['rank'] = happiness_rank['rank'].astype(int)
happiness_rank
```

```python
# Happiness rank by year (sorted by 2017)

whr_all_years.pivot_table(index='country', columns='year', values='happiness_rank').sort_values(by=2017)
```

```python
# Countries with least trust in government

whr_all_years.pivot_table(index='country', values='trust_government_corruption').sort_values(by='trust_government_corruption')
```

```python
# Countries with most trust in government

whr_all_years.pivot_table(index='country', values='trust_government_corruption').sort_values(by='trust_government_corruption', ascending=False)
```
